```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
df = pd.read_csv('archive/acs2017_census_tract_data.csv')
# df.head()
# df.describe()
df.columns
```

```python
incomespc = df['IncomePerCap']
high_income = pd.Series([income for income in incomespc if income >= 42_000])

med_income = pd.Series([income for income in incomespc if income >= 25_000 and income < 42_000])

low_income = pd.Series([income for income in incomespc if income < 25_000])

print(len(high_income), len(med_income), len(low_income))

# low_income.describe()

# plt.hist([high_income, med_income, low_income])

df['high_income'] = df['IncomePerCap'] >= 42_000
df['med_income'] = df['IncomePerCap'] >= 25_000
df['med_income'][df['IncomePerCap'] >= 42_000] = False
df['low_income'] = df['IncomePerCap'] < 25_000
df['IncomeRange'] = ""
df['IncomeRange'][df['low_income']] = 'low'
df['IncomeRange'][df['med_income']] = 'med'
df['IncomeRange'][df['high_income']] = 'high'


# print(df[['high_income', 'med_income', 'low_income', 'IncomeRange','IncomePerCap']].sample(100))
sns.scatterplot(x=df['Professional'], y=df['IncomePerCap'], hue=df['IncomeRange'])
```
