```python
for col in df.columns:
    try:
        corrs = df.corr()
        corrs['AbsoluteCorrelation'] = abs(corrs[col])
        print(col, "\n", corrs.sort_values(by='AbsoluteCorrelation', ascending=False).head(10)[col], "\n\n")
    except:
        continue
```
